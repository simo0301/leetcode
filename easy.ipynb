```python
#13(easy)
class Solution:
    def romanToInt(self, s):
        dict = {'I':1, 'V':5, 'X':10, 'L':50, 'C':100, 'D':500, 'M':1000}
        sum = 0
        i = 0    
        while i < len(s)-1:
            if dict[s[i]] < dict[s[i+1]]:
                sum = sum - dict[s[i]]

            else:
                sum = sum + dict[s[i]]
            i+=1
        sum = sum + dict[s[-1]]
        return sum
    
s = "IV"
Solution().romanToInt(s)   

```

```python
#14(easy)
strs = ["flower","flow","flight"]
class Solution:
    def longestCommonPrefix(self, strs):
        min = len(strs[0])
        a = 0
        for i in strs:
            if len(i) < min:
                a = strs.index(i)
                min = len(i)
        s = strs[a]
        initial = ""
        for j in range(len(s)):
            for i in strs:
                if s[j] != i[j]:
                    return initial
            else:
                initial+=s[j]
        return initial
Solution().longestCommonPrefix(strs)
```
